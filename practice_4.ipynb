The dataset “Supermart Grocery Sales.csv” represents sales data from a grocery store, capturing various aspects of transactions including the order ID, customer name, product category and subcategory, city and region of the order, order date, sales amount, discount applied, and profit generated. Each row corresponds to a unique transaction, providing insights into customer buying patterns, product popularity, and sales performance across different geographical regions.

Create appropriate visualizations to answer the given questions.

Question 1:
Create a histogram to visualize the distribution of sales amounts in the dataset. Set appropriate bin widths and labels for the x-axis and y-axis. Title the histogram as "Distribution of Sales Amounts".

```r
# Load the dataset
sales_data <- read.csv("supermart_sales.csv")

# Create a histogram for sales amounts
hist(sales_data$Sales,
     breaks = 20,
     col = "blue",
     main = "Distribution of Sales Amounts",
     xlab = "Sales Amount",
     ylab = "Frequency")
```

Question 2:
Create a pie chart to represent the proportion of total sales contributed by each region (North, South, West). Provide a legend for the regions and title the pie chart as "Sales Distribution by Region".

```r
head(sales_data)
```

```r
# Aggregate sales by region
sales_by_region <- aggregate(Sales ~ Region, data = sales_data, sum)

# Create a pie chart
pie(sales_by_region$Sales,
    labels = sales_by_region$Region,
    main = "Sales Distribution by Region",
    col = rainbow(length(sales_by_region$Sales)))
```

Question 3:
Create a line chart to visualize the trend of total sales over time. Group the sales data by year, and plot the yearly total sales on the y-axis against time on the x-axis. Add gridlines and title the line chart as "Total Sales by Year".

```r
library(tidyverse)

# Aggregate total sales by year
sales_by_year <- sales_data %>%
  group_by(Year) %>%
  summarise(Total_Sales = sum(Sales))

# Create a line graph for total sales by year
ggplot(sales_by_year, aes(x = Year, y = Total_Sales)) +
  geom_line() +
  geom_point() +
  labs(title = "Total Sales by Year",
       x = "Year",
       y = "Total Sales") +
  theme_minimal()
```

Question 4:
Create a bar chart to compare the average sales amounts across different product categories. Arrange the categories on the x-axis and the average sales on the y-axis. Provide labels for each axis and title the bar chart as "Average Sales by Category".

```r
# Aggregate average sales by category
avg_sales_by_category <- aggregate(Sales ~ Category, data = sales_data, mean)

# Create a bar chart
barplot(avg_sales_by_category$Sales,
        names.arg = avg_sales_by_category$Category,
        col = "blue",
        main = "Average Sales by Category",
        xlab = "Category",
        ylab = "Average Sales")
```

Question 5:
Create a scatter plot to explore the relationship between sales and profit. Plot sales amounts on the x-axis and profit on the y-axis. Add a trendline to the scatter plot and title it as "Sales vs. Profit".

```r
# Create a scatter plot for Sales vs. Profit
plot(sales_data$Sales, sales_data$Profit,
     main = "Sales vs. Profit",
     xlab = "Sales",
     ylab = "Profit",
     pch = 19,
     col = "blue")

# Add a trendline
abline(lm(Profit ~ Sales, data = sales_data), col = "red")
```
